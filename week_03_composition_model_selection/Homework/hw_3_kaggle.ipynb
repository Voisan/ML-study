Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
#Сделаем LabelEncoding,поскольку тут много признаков с двумя уникальными значениями
pat = {
    "No": 0,
    "Yes": 1,
    "No internet service": 0,
    "No phone service": 0,
    "Male": 0,
    "Female": 1,
    "DSL": 1,
    "Fiber optic": 2,
    "Month-to-month": 0,
    "One year": 1,
    "Two year": 2,
    "Credit card (automatic)": 0,
    "Bank transfer (automatic)": 1,
    "Mailed check": 2,
    "Electronic check": 3,
}
data.replace(pat,inplace=True)
dtest.replace(pat,inplace=True)
```

```python
y = data["Churn"]
X = data.drop(columns=["Churn"])
```

```python
#Разделим на тренировычные и тестовые части
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=102)
# Масштабирование данных
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
dtest_scaled = scaler.transform(dtest)
```

```python
# Создание модели
logres = LogisticRegressionCV(Cs=20, cv=10, scoring="roc_auc", random_state=52, refit=True, max_iter=1000)

logres.fit(X_train_scaled, y_train)
auc = logres.score(X_test_scaled, y_test)
print("Лучшие параметры:", logres.get_params())
print(auc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=52)

# Параметры для поиска
param_grid = {
    'n_estimators': [1000],
    'max_depth': [10,15],
    'min_samples_split': [10,15],
    'min_samples_leaf': [5,10]
}

# Поиск лучших параметров с кросс-валидацией
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring="roc_auc")
grid_search.fit(X_train_scaled, y_train)
```

```python
# Лучшие параметры
print("Лучшие параметры:", grid_search.best_params_)

# Оценка на тестовых данных
best_model = grid_search.best_estimator_
auc_test = best_model.score(X_test_scaled, y_test)
print(f"AUC-ROC на тестовых данных: {auc_test}")
```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметров количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию.

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
from catboost import CatBoostClassifier, Pool
#Воспользуемся моделью CatBoostClassifier
catboost = CatBoostClassifier(
    cat_features=cat_cols,
    logging_level="Silent",
    eval_metric="AUC:hints=skip_train~false",
    grow_policy="Lossguide",
    metric_period=1000,
    random_seed=0,
)
```

```python
#Зададим параметры для перебора
par = {
    "n_estimators": [100,150,200,250],
    "learning_rate": [0.01,0.05,0.1],
    "max_depth": [1,4,10],
    "l2_leaf_reg": [5,10],
    "subsample": [0.1,0.3,0.6],
    "random_strength": [1,5,10],
    "min_data_in_leaf": [100,150,200]
}
```

```python
# Создание пула данных
train_pool = Pool(X_train, label=y_train, cat_features=cat_cols)

# Поиск лучших параметров с кросс-валидацией
catboost.grid_search(par, train_pool, cv=5, plot=True, refit=True)
```

```python
print("Best score:", end=' ')
print(catboost.best_score_)
best_params = catboost.get_params()
for f in ("cat_features", "logging_level", "eval_metric"):
    best_params.pop(f)
print("Best params:", end=' ')
print(best_params)
```
